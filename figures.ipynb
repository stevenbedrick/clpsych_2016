```python
import pynini as pyn
```

```python
first_part = pyn.acceptor("[I can't move my]")
first_part
```

```python
sigma = pyn.optimize(pyn.union(*["[{}]".format(w) for w in ["money","big","...","upper","lower", "right","left"]]))
sigma
```

```python
lattice = pyn.rmepsilon(first_part + sigma + pyn.acceptor("[hand]"))
```

```python
lattice
```

```python
lattice.draw("lattice.dot", acceptor=True, portrait=True)
```
